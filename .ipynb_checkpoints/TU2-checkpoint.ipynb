```python
url = "https://sports.sindonews.com/read/1235354/51/daftar-lengkap-perolehan-medali-sea-games-2017-1504095641"
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')

findtable =str(soup.find_all('table'))
soup = BeautifulSoup(findtable,'html.parser' )
indexpos = [6]
pos = 6
datapos = []

indexnegara = [7]
negara = 7
datanegara=[]

indexemas = [8]
emas = 8
dataemas =[]

indexperak = [9]
perak = 9
dataperak =[]

indexperunggu = [10]
perunggu = 10
dataperunggu = []

indextotal = [11]
total = 11
datatotal = []
findvalue = soup.find_all('td')

for i in findvalue :
    pos += 6
    negara += 6
    emas += 6
    perak += 6
    perunggu += 6
    total += 6
    indexpos.append(pos)
    indexnegara.append(negara)
    indexemas.append(emas)
    indexperak.append(perak)
    indexperunggu.append(perunggu)
    indextotal.append(total)

for i in range(len(findvalue)) :
    if i in indexpos :
        datapos.append(findvalue[i].text)
    elif i in indexnegara :
        datanegara.append(findvalue[i].text)
    elif i in indexemas :
        dataemas.append(findvalue[i].text)
    elif i in indexperak :
        dataperak.append(findvalue[i].text)
    elif i in indexperunggu :
        dataperunggu.append(findvalue[i].text)
    elif i in indextotal :
        datatotal.append(findvalue[i].text)

data = []
for i in range(len(datapos)) :
    data.append({
        'pos' : datapos[i] ,
        'negara' : datanegara[i] ,
        'emas' : dataemas[i] ,
        'perak' : dataperak[i] ,
        'perunggu' : dataperunggu[i] ,
        'total' : datatotal[i]
    })

df= pd.DataFrame(data)
df['emas'] = pd.to_numeric(df['emas'])
df = df.sort_values('negara')
df
```

### Soal 2 - SPLTV (Sistem Persamaan Linear Tiga Variabel)

```python
a = np.array ([
    [1,-2,1],
    [3,1,-2],
    [7,-6,-1]
])
b = np.array([6,4,10])
x,y,z = np.linalg.solve(a,b)[0],np.linalg.solve(a,b)[1],np.linalg.solve(a,b)[2]

print('nilai x = ',round(x))
print('nilai y = ', round(y))
print('nilai z = ', round(z))
```

```python
fig = plt.figure(figsize = (18,6))

axes1 = plt.subplot(131, projection='3d')
x1 = np.array([0, 0, 6])
y1 = np.array([-3, 0, 0])
z1 = np.array([0, 6, 0])
axes1.scatter(x1, y1, z1)
coords_1 = [list(zip(x1, y1, z1))]
surface_1 = Poly3DCollection(coords_1, alpha=.25, facecolor='r')
axes1.add_collection3d(surface_1)
axes1.set_xlabel('nilai X')
axes1.set_ylabel('nilai Y')
axes1.set_zlabel('nilai Z')

axes2 = plt.subplot(132, projection='3d')
x2 = np.array([0.6, 0, 0])
y2 = np.array([4, 4, 0])
z2 = np.array([-1.5, 0, -2])
axes2.scatter(x2, y2, z2)
coords_2 = [list(zip(x2, y2, z2))]
surface_2 = Poly3DCollection(coords_2, alpha=.25, facecolor='g')
axes2.add_collection3d(surface_2)
axes2.set_xlabel('nilai X')
axes2.set_ylabel('nilai Y')
axes2.set_zlabel('nilai Z')


axes3 = plt.subplot(133, projection='3d')
x3 = np.array([0, 0, 1.4])
y3 = np.array([-1.67, 0, 0])
z3 = np.array([0, -10, 0])
coords_3 = [list(zip(x3, y3, z3))]
surface_3 = Poly3DCollection(coords_3, alpha=.25, facecolor='b')
axes3.add_collection3d(surface_3)
axes3.scatter(x3, y3, z3)
axes3.set_xlabel('nilai X')
axes3.set_ylabel('nilai Y')
axes3.set_zlabel('nilai Z')

plt.show()
```

```python
dfBcg = pd.read_csv('Balita Terimunisasi BCG 1995-2017.csv',na_values=['n.a'])
dfCampak = pd.read_csv('Balita Terimunisasi Campak 1995-2017.csv',na_values=['n.a'])
dfDpt = pd.read_csv('Balita Terimunisasi DPT 1995-2017.csv',na_values=['n.a'])
dfPolio = pd.read_csv('Balita Terimunisasi Polio 1995-2017.csv',na_values=['n.a'])

# rename kolom agar nama lebih simpel
dfBcg = dfBcg.rename(columns={'% Balita yang pernah mendapat imunisasi BCG':'%_imunisasi_BCG'})
dfCampak = dfCampak.rename(columns={'% Balita yang pernah mendapat imunisasi Campak':'%_imunisasi_campak'})
dfDpt = dfDpt.rename(columns={'% Balita yang pernah mendapat imunisasi DPT':'%_imunisasi_DPT'})
dfPolio = dfPolio.rename(columns={'% Balita yang pernah mendapat imunisasi Polio':'%_imunisasi_polio'})

# interpolasi kolom persentase
dfBcg = dfBcg.interpolate(axis=0)
dfCampak = dfCampak.interpolate(axis=0)
dfDpt = dfDpt.interpolate(axis=0)
dfPolio = dfPolio.interpolate(axis=0)

# buat figur persentase balita yang sudah diimunisasi
fig1, ax1 = plt.subplots(2, 2, figsize=(11,6))
fig1.tight_layout()
fig1.subplots_adjust(hspace=0.4)
ax1[0,0].bar(dfBcg.index, dfBcg['%_imunisasi_BCG'], color='r')
ax1[0,0].set_title('BCG', size=10)
ax1[0,1].bar(dfCampak.index, dfCampak['%_imunisasi_campak'], color='g')
ax1[0,1].set_yticks(np.arange(0,81,10))
ax1[0,1].set_title('Campak', size=10)
ax1[1,0].bar(dfDpt.index, dfDpt['%_imunisasi_DPT'], color='y')
ax1[1,0].set_title('DPT', size=10)
ax1[1,1].bar(dfPolio.index, dfPolio['%_imunisasi_polio'], color='b')
ax1[1,1].set_title('Polio', size=10)

for (row, col), subplot in np.ndenumerate(ax1) :  # set x tick biar gak usah satu-satu
    subplot.set_xticks(dfBcg.index)
    subplot.set_xticklabels(dfBcg.index, rotation=90)
    
# buat figur persentasi balita yang belum diimunisasi
fig2, ax2 = plt.subplots(2, 2, figsize=(11,6))
fig2.tight_layout()
fig2.subplots_adjust(hspace=0.4)
ax2[0,0].bar(dfBcg.index, 100 - dfBcg['%_imunisasi_BCG'], color='r')
ax2[0,0].set_title('BCG', size=10)
ax2[0,1].bar(dfCampak.index, 100 - dfCampak['%_imunisasi_campak'], color='g')
ax2[0,1].set_title('Campak', size=10)
ax2[1,0].bar(dfDpt.index, 100 - dfDpt['%_imunisasi_DPT'], color='y')
ax2[1,0].set_title('DPT', size=10)
ax2[1,1].bar(dfPolio.index, 100 - dfPolio['%_imunisasi_polio'], color='b')
ax2[1,1].set_title('Polio', size=10)

for (row, col), subplot in np.ndenumerate(ax2) :  # set x tick biar gak usah satu-satu
    subplot.set_xticks(dfBcg.index)
    subplot.set_xticklabels(dfBcg.index, rotation=90)
```
